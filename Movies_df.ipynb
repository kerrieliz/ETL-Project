```python
# Setting up our url's
discover = 'https://api.themoviedb.org/3/discover/movie?'
movie_url = 'https://api.themoviedb.org/3/movie/'
rating = 'http://www.omdbapi.com/?t='
```

```python
# Movie year range
movie_years = np.arange(2008,2019,1)
movie_years
```

```python
# Empty arrays to store our data
budget = []
revenue = []
popularity = []
runTime = []
voteAve = []
genre = []
prodComp = []
releaseDate = []
voteCount = []
title = []
movieID = []
ratings = []
```

```python
# Loop through our years
for year in movie_years:
    f_url = discover + "api_key=" + str(api_key) + "&language=en-US&primary_release_year="+str(year)+"&sort_by=popularity.desc"
    data = requests.get(f_url).json()
    try:
        for movie in data['results']:
            try:
                title.append(movie['title'])
                movieID.append(movie['id'])
                print(f"Gotten {movie['title']}!")
            except:
                print("Something went wrong, oops!")
    except:
        print("Something seems to have gone wrong with the fetching.")
```

```python
api_key3 ='&apikey=' + api_key2
for ti in title:
    urlr = rating + ti + api_key3
    response = requests.get(urlr).json()
    try:
        ratings.append(response['Rated'])
        print(f'Rating from {ti}.')
    except:
        print(f'No Rating for {ti}.')
        ratings.append('none')
```

```python
for i in movieID:
    url2 = movie_url + str(i) +'?api_key=' + str(api_key) +"&language=en-US"
    data2 = requests.get(url2).json()
    try:
        budget.append(data2['budget'])
        revenue.append(data2['revenue'])
        popularity.append(data2['popularity'])
        runTime.append(data2['runtime'])
        voteAve.append(data2['vote_average'])
        genre.append(data2['genres'])
        prodComp.append(data2['production_companies'])
        releaseDate.append(data2['release_date'])
        voteCount.append(data2['vote_count'])
        print(f"More Data on {data2['title']}")
    except:
        print(f'Something went wrong on id {i}!')
```

```python
tit2 = title.copy()
del title[0:13]
movID = movieID.copy()
del movieID[0:13]
rat= ratings.copy()
del ratings[0:13]
```

```python
print(len(budget), 
      len(revenue),
      len(popularity), 
      len(runTime), 
      len(voteAve), 
      len(genre),
      len(prodComp),
      len(releaseDate),
      len(voteCount),
      len(title),
      len(movieID ),
      len(ratings))
```

```python
movie_info = {
    'Title' : title,
    'ID' : movieID,
    'Budget': budget,
    'Revenue': revenue,
    'Popularity': popularity,
    'Run_Time': runTime,
    'Vote_Average': voteAve,
    'Vote_Count': voteCount,
    'Genre': genre,
    'Production_Company': prodComp,
    'Release_Date': releaseDate,
    'Rating' : ratings
}
movie_df = pd.DataFrame(movie_info)
movie_df.head()
```

```python
file = "CSV_Files\Movies.csv"

new_df = pd.read_csv(file,encoding="ISO-8859-1")
new_df.head()
```

```python
genre = []
title = []
revenue = []
budget = []
rating = []
popularity = []
vote_count = []
vote_avg = []


# iterating through to split the string for Genre
for i,row in new_df.iterrows():
    string = new_df['Genre'][i].split(',')
#     print(string)

# Iterating through the string and partitioning to get to name and id of genre
    for genstring in string:
        cutStr = genstring.partition(': ')
#         print(cutStr[0])
# If the first part of each cutStr partition is = name then exract the information
        if cutStr[0] == " 'name'":
#         To get the genre name, we need to look at the second part of the cutStr and extract the actual genre without characters
            genre.append(cutStr[2][1:-2])
            title.append(row['Title'])
            revenue.append(row['Revenue'])
            budget.append(row['Budget'])
            rating.append(row['Rating'])
            popularity.append(row['Popularity'])
            vote_count.append(row['Vote_Count'])
            vote_avg.append(row['Vote_Average'])
            
genre_df = pd.DataFrame({"Genre": genre,
                        "Title": title,
                        "Revenue": revenue,
                        "Budget": budget,
                        "Rating": rating,
                        "Popularity": popularity,
                        "Vote_Count": vote_count,
                        "Vote_Average": vote_avg
                        })
genre_df.head()
```

```python
genre_df['Genre']=genre_df['Genre'].replace('\'','',regex=True)
genre_df.head()
```

```python
genre_df.to_csv('Genre.csv',index=False)
```

```python
genre_df.sort_values('Revenue',ascending=False)
new_df = genre_df.drop_duplicates(['Title'],keep='first')
new_df = new_df.sort_index()
new_df
```
