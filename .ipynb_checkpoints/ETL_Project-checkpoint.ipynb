```python
import pandas as pd 
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
#import numpy as np
```

```python
movie_file = "Genre2.csv"
movie_df = pd.read_csv(movie_file)
movie_df.head()
```

```python
vg_file = "Video_Games_Sales_as_at_22_Dec_2016.csv"
vg_df = pd.read_csv(vg_file)
vg_df.head()
```

```python
# Create filtered dataframe with desired columns 
movie_cols = ["Title", "Genre", "Revenue"]
movie_transformed = movie_df[movie_cols].copy()

# Rename columns
movie_transformed = movie_transformed.rename(columns={"Title": "movie_title",
                                                     "Genre": "genre",
                                                    "Revenue": "sales"})

# Clean data - drop dupes and set index 
movie_transformed.drop_duplicates("movie_title", inplace=True)
movie_transformed.set_index("genre", inplace=True)

movie_transformed.head()
```

```python
# Create filtered dataframe with desired columns 
vg_cols = ["Name", "Year_of_Release", "Genre", "NA_Sales"]
vg_transformed = vg_df[vg_cols].copy()

# Rename columns
vg_transformed = vg_transformed.rename(columns={"Name": "video_game_name",
                                                "Year_of_Release": "year",
                                                "Genre" : "genre",
                                                "NA_Sales": "sales"})

# Clean data - drop dupes and set index 
vg_transformed.drop_duplicates("video_game_name", inplace=True)
vg_transformed.set_index("genre", inplace=True)

vg_transformed.head()
```

```python
# postgresql://user:password@localhost:5432/database_name

connection_string = "postgres:postgres@localhost:5432/genres_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
# Load dataframes into database 
movie_transformed.to_sql(name='movies', con=engine, if_exists='append', index = True)
```

```python
vg_transformed.to_sql(name='video_games', con=engine, if_exists='append', index = True)
```

```python
# Visuals 
movie_df2 = movie_df[["Genre" , "Revenue"]]
movie_groupby = movie_df2.groupby(['Genre']).sum()
movie_groupby.head()
```

```python
movie_groupby.plot(kind='bar')
movie_groupby
plt.title("Movie sales in North America by Genre")
plt.ylabel("# of sales in millions")
plt.xlabel("Genre")
plt.savefig("Movie_sales")
plt.show()
```

```python
vg_df2 = vg_df[["Genre" , "NA_Sales"]]
vg_groupby = vg_df2.groupby(['Genre']).sum()
vg_groupby.head()
```

```python
vg_groupby.plot(kind='bar')
vg_groupby
plt.title("Video Game sales in North America by Genre")
plt.ylabel("# of sales in millions")
plt.xlabel("Genre")
plt.savefig("VGsales")
plt.show()
```
