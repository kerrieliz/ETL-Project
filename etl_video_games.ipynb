```python
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
VG_df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
```

```python
VG_df
```

```python
VG_df2 = VG_df[["Genre" , "NA_Sales"]]
VG_df2
```

```python
VG_groupby = VG_df2.groupby(['Genre']).sum()
VG_groupby
```

```python
VG_groupby.plot(kind='bar')
VG_groupby
plt.title("Video Game sales in North America by Genre")
plt.ylabel("# of sales in millions")
plt.xlabel("Genre")
plt.savefig("VGsales")
plt.show()
```
